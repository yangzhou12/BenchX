```python
import pandas as pd
```

```python
df = pd.read_csv('training.csv', index_col=0)
df.head()
```

```python
# Extract dicom_id
df['dicom_id'] = df['image_path'].apply(lambda x: x.split('/')[4][:-4])
df.head()
```

```python
# Rename column image_path to Path
df = df.rename(columns={'image_path': 'Path'})
df.head()
```

```python
split_df = pd.read_csv('mimic-cxr-2.0.0-split.csv')
split_df.head()
```

```python
df = df.merge(split_df, how='inner', on='dicom_id')
df.head()
```

```python
df['split'] = df['split'].apply(
        lambda x: "valid" if x == "validate" or x == "test" else x)
df.head()
```

```python
# Read metadata csv file
metadata_df = pd.read_csv('mimic-cxr-2.0.0-metadata.csv')
metadata_df.head()
```

```python
# Merge metadata df with df
df = df.merge(metadata_df, how='inner', on='dicom_id')
```

```python
# Filter frontal views only
df = df[df['ViewPosition'].isin(["PA", "AP"])]
df.head()
```

```python
len(df) # number of image-text pairs after filtering for frontal views
```

```python
import re
```

```python
def parse_report(report, key):
    matches = list(re.compile(r'(?P<title>[ [A-Z\]()]+):').finditer(report)) #Remove carat and newline flag
    parsed_report = {}
    for (match, next_match) in zip(matches, matches[1:] + [None]):
        start = match.end()
        end = next_match and next_match.start()

        title = match.group('title')
        title = title.strip()

        paragraph = report[start:end]
        paragraph = re.sub(r'\s{2,}', ' ', paragraph)
        paragraph = paragraph.strip()

        parsed_report[title] = paragraph.replace('\n', '\\n')

    if key in parsed_report:
        if parsed_report[key] != '':
            return parsed_report[key]
    return " "
```

```python
df['findings'] = df['report_content'].apply(lambda x: parse_report(x, 'FINDINGS'))
df['impression'] = df['report_content'].apply(lambda x: parse_report(x, 'IMPRESSION'))
```

```python
master_df = df.loc[:, ['Path', 'dicom_id', 'split', 'findings', 'impression']]
master_df.head()
```

```python
master_df['findings'] = df["findings"].map(str)
master_df['impression'] = df["impression"].map(str)
```
